```python
import numpy as np
def normalize(A):
    dim1, dim2 = A.shape[0], A.shape[1]
    
    min = A.min()
    max = A.max()
    
    for i in range(dim1):
        for j in range(dim2):
            x = A[i,j]
            z = (x-min)/(max-min)
            
            A[i,j] = z
            
    return A
```

```python
A = np.random.normal(0,10,(10,10))
```

```python
normalize(A)
```
